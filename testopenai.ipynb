```python
# Example dummy function hard coded to return the same weather
# In production, this could be your backend API or an external API
import requests
def get_current_weather(location):
    """Get the current weather in a given location"""

    url = "https://ai-weather-by-meteosource.p.rapidapi.com/find_places"

    querystring = {"text":location,"language":"en"}

    headers = {
        "X-RapidAPI-Key": "f1e6cf5258msh554488bd2db6449p1bda98jsnc5546906f82e",
        "X-RapidAPI-Host": "ai-weather-by-meteosource.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    print(response.json())
    return response.json()
```

```python
functions = [
        {
            "name": "get_current_weather",
            "description": "Get the current weather in a given location",
            "parameters": {
                "type": "object",
                "properties": {
                    "location": {
                        "type": "string",
                        "description": "The city and state, e.g. San Francisco, CA",
                    },
                    
                },
                "required": ["location"],
            },
        }
    ]
```

```python
user_message="What is the temperature of Bangalore"

messages.append({"role": "user", "content": user_message})
completion=openai.ChatCompletion.create(
    model="gpt-3.5-turbo-0613",
    messages=messages,
    functions=functions
)
```

```python
completion
```

```python
print(completion.choices[0].message)
```

```python
response=completion.choices[0].message
```

```python
response
```

```python
function_name=response['function_call']['name']
print(function_name)
```

```python
import json
location=eval(response['function_call']['arguments'])['location']
print(location)
```

```python
# Step 4: send the info on the function call and function response to GPT
messages.append(response)  # extend conversation with assistant's reply
messages.append(
    {
        "role": "function",
        "name": function_name,
        "content": location,
    }
)
```

```python
messages
```

```python
# extend conversation with function response
second_response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo-0613",
    messages=messages,
    functions=functions
)  # get a new response from GPT where it can see the function response
```

```python
print(second_response.choices[0].message)
```
